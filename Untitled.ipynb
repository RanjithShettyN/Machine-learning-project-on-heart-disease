```python
import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
np.random.seed(45)
#read the csv data

heart = pd.read_csv("heart-disease.csv")

#seperate the target and the other parameters

m = heart.drop("target", axis= 1)
n = heart["target"]

#now split the data into testing and training

m_train, m_test, n_train, n_test = train_test_split(m,n,test_size = 0.2)

#now train the model on the traing datdset

model1 = RandomForestClassifier()
model1.fit(m_train,n_train)

#now check the performance of the model

model1.score(m_test,n_test)
```

```python
from sklearn.metrics import roc_curve


y_proba1 = model1.predict_proba(m_test)

y_proba1[:10]
```

```python
y_probs_positive = y_proba1[:,1]

y_probs_positive[:5]
```

```python
fpr, tpr, threshholds = roc_curve(n_test, y_probs_positive)
```

```python
fpr
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_roc_curve(fpr, tpr):
    """
    plots the roc curve give the false positive rate and the true positive rate of the given model
    
    """
    plt.plot(fpr,tpr,color = "orange", label = "ROC")
    plt.plot([0,1],[0,1], color = "blue", linestyle = "--", label = "guessing")
    plt.xlabel("False positive Rate (fpr)")
    plt.ylabel("True positive rate (tpr)")
    plt.title("Receiver Operating characteristic curve (ROC) ")
    plt.show()
```

```python
plot_roc_curve(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(n_test, y_probs_positive)
```

## Confusion Marix

a confusion matrix is a quick way to compare the label a model predicts and the actual label it was supposed to predict

in essence giving you an idea where the model is getting confused


```python
from sklearn.metrics import confusion_matrix

y_preict = model1.predict(m_test)

confusion_matrix(n_test, y_preict)
```

```python
pd.crosstab(n_test,
            y_preict,
            rownames = ["Actual labels"],
            colnames = ["predicted labels"])
```

```python
import seaborn as sns

sns.set(font_scale = 1.5)

conf_mat = confusion_matrix(n_test, y_preict)

sns.heatmap(conf_mat);
```
